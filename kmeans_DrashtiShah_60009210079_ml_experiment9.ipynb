```python
# Calculate SSE for different number of clusters
sse_values = []
for n_clusters in range(2, 11):
    kmeans.fit(points)
    sse = kmeans.inertia_
    sse_values.append(sse)
```

```python
# Plot SSE values against number of clusters (elbow method)
plt.plot(range(2, 11), sse_values, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
